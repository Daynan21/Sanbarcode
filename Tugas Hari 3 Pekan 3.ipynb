```python
# code here
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
fig, ax = plt.subplots(ncols=3, figsize=(12,8))
ax[0].plot(x, tan, color='b', label='tanX')
ax[1].plot(x, cos, color='r', label='sinX')
ax[2].plot(x, sin, color='g', label='cosX')
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x, y, color='blue')
ax.set_title('Grafik CosX')
ins1= ax.inset_axes([0.8, 0.8, 0.4, 0.4])
ins1.plot(x, y, color='blue')
ins1.plot(x, y4, color='black')
ins1.set_title('Grafik CosX^4')
ins2= ax.inset_axes([-0.1, -0.1, 0.4, 0.4])
ins2.plot(x, y, color='blue')
ins2.plot(x, y2, color='red')
ins2.set_title('Grafik CosX^2')
ins3= ax.inset_axes([-0.1, 0.8, 0.4, 0.4])
ins3.plot(x, y, color='blue')
ins3.plot(x, y5, color='yellow')
ins3.set_title('Grafik CosX^5')
ins4= ax.inset_axes([0.8, -0.1, 0.4, 0.4])
ins4.plot(x, y, color='blue')
ins4.plot(x, y3, color='green')
ins4.set_title('Grafik CosX^3')
plt.show()
```
